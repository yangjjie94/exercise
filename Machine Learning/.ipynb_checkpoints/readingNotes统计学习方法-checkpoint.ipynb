## 前言
这本书的目的在于教会我们利用数据构建传统机器学习模型及利用模型对数据进行分析与预测。而不是数据的观测和收集。统计学习方法中又以监督学习方法尾注，主要包括用于**分类**、**标注**与**回归**问题的方法，这些方法在*自然语言处理、信息检索、文本数据挖掘等领域*中有着极其广泛的应用。  

## 第一章 <a name="chap1"></a>
### 1.2 统计学习
统计学习（statistical learning），又称统计机器学习（statistical machine learning），是关于计算机技术数据构建概率统计模型并运用模型对数据进行预测与分析的一门学科。

1\. 其特点是
  
  
- 建立在计算机及网络平台上，不是paperwork，也不是人的计算力能达到的
- 数据驱动。以数据为研究对象，可以类比理化生以实验为基础，数据就是机器学习的上限
- 目的是对数据进行预测与分析。分析面向过去，预测面向未来。
- 统计学习是概率论、统计学、信息论、计算理论、最优化理论及计算机科学等多个领域的交叉学科，并且在发展中逐步形成独自的理论体系与方法论。

Herbert A. Simon曾定义“学习”：“如果一个系统能够通过执行某个过程改进它的性能，这就是学习”。

2\. 统计学习的对象
数据，数据，数据！数据是指存在与计算机以网络上的各种数字、文字、图像、音频、视频。统计学习从数据触发，提取数据的特征，出相处数据的模型，发现数据的知识，又回到对数据的分析与预测中去。  
统计学习关于数据的基本假设是同类数据具有一定的统计规律性。统计规律性意味着可以用概率通常方法来加以处理，比如，可以用随机变量描述数据中的特征，用概率分布描述数据的通常规律。  
我们用变量或变量组表示数据。
数据分为连续变量和离散变量。

3\. 统计学习的目的
预测与分析

4\. 统计学习的方法
基于数据构建统计模型从而对数据进行预测与分析。

**统计分析**包括**监督学习（supervised learning)**、**非监督学习（unsupervised learning）**、**强化学习（reinforcement learning)**等部分。  
从监督学习的角度，我们可以进一步把统计学习的方法定义为：
从给定的、有限的、由于学习的训练数据（training data）集合出发，假设数据是独立同分布产生的；并且假设要学习的模型属于某个函数的集合，称为**假设空间（hypothesis space）**；应用某个**评价标准（evaluation criterion）**，从假设空间中选取一个最优的模型，使它对已知训练数据以及未知数据（test data）在给定的评价准则下有最优的预测；最优模型的选取由**模型学习的算法**实现。  
也称为统计学习三要素，简称模型（model）、策略（strategy）和算法（algorithm）

实现统计学习方法的步骤如下：

1. 得到一个有限的训练**数据集**合；
2. 确定包含所有可能的模型的**假设空间**，即学习模型的集合；
3. 确定模型选择的准则，即学习的**策略**；
4. 实现求解最优模型的算法，即学习的**算法**；
5. 通过学习方法选择最优模型；
6. 利用学习的最有模型对新数据进行预测或分析；

5\. 统计学习的重要性
- 处理海量数据的有效手段
- 计算机智能化的有效方法
- 计算机科学发展（系统、计算、信息三位一体）的一个重要组成部分（信息）



### 1.2 监督学习
监督学习，任务是学习一个模型，使模型能够对任意给定的输入，对其相应的输出做出一个好的预测。
#### 1.2.1 基本概念
- 输入空间input space: 输入所有可能取值的集合
- 输出空间output space: 输出所有可能取值的集合
> 输入输出空间可以是有限或无限，可以是同一个空间也可以不是；但通常输出空间小于输入空间。<br>
- 实例instance: 每个具体的输入，通常由特征向量feature vector表示。
- 特征空间feature space: 所有特征向量存在的空间，特征空间的每一维对应于一个特征，模型都是定义在特征空间上的。
> 特征空间和输入空间有时候是一个，不作区分；有时输入空间与特征空间为不同的空间。  
- 样本：输入输出对，输入（或特征向量）$X$和输出$Y$打包成训练集$T={(x_1,y_1),(x_2,y_2),...,(x_N,y_N)}$  
> 根据输入变量$X$和$Y$是连续还是离散可以把预测问题分为`*回归问题*`、`*标注问题*`和`*分类问题*`
- 回归问题：$X$连续，$Y$连续
- 标注问题：$X$变量序列，$Y$变量序列
- 分类问题：$Y$为有限个离散变量

- 假设空间：模型是一个由输入空间到输出空间的映射，这样的模型的集合就是假设空间。假设空间的确定意味着学习范围的确定。
监督学习的模型可以是概率模型或非概率模型，由条件概率分布$P(Y|X)$或决策函数（decision function） $Y=f(X)$表示，随具体学习方法而定。

1.2.2 问题的形式化
监督学习分为**学习**和**预测**两个过程，由**学习系统**和**预测系统**完成。  
假设训练数据与测试数据是依联合概率分布$P(X,Y)$ **独立同分布**产生的。  

- 学习系统：学习得到的模型描述输入与输出随机变量之间的映射关系，表示为条件概率分布$\hat{P}(Y|X)$或决策函数$Y=\hat{f}(X)$。
- 预测系统：对于给定的输入$x_{N+1}$，由模型$y_{N+1}=arg max \hat{P}(y_{N+1}|y_{N})$或$y_{N+1}=\hat{f}(x_{N+1})$

一个好的预测就是$y_i$和$f(x_i)$差足够小，不仅对训练数据集有足够好的预测，同时对位置的测试数据集的预测也有尽可能好的推广。

### 1.3 统计学习三要素
统计学习方法 = 模型 + 策略 + 算法
#### 1.3.1 模型
模型就是所要学习的条件概率分布或决策函数。要从假设空间中的所有条件概率分布或决策函数中找到一个。

- 决策分布的集合——非概率模型

假设空间hypothesis space用$\mathscr{F}=\{f|Y=f(X)\}$表示，其中$X$和$Y$是定义在输入空间$\mathscr{X}$和输出空间$\mathscr{Y}$上的变量。这时$\mathscr{F}$通常是由一个参数向量决定的函数族：
$\mathscr{F}=\{f|Y=f_0(X),\theta\in\mathbb{N}^n\}$
参数向量$\theta$取值于$n$维欧式空间，称为参数空间。

- 条件概率的集合——概率模型

假设空间hypothesis space定义为条件概率的集合  
$\mathscr{F}=\{P|P(Y|X)\}$  
，其中$X$和$Y$是定义在输入空间$\mathscr{X}$和输出空间$\mathscr{Y}$上的变量。这时$\mathscr{F}$通常是由一个参数向量决定的函数族：
$\mathscr{F}=\{P|P(Y|X),\theta\in\mathbb{N}^n\}$
参数向量$\theta$取值于$n$维欧式空间，称为参数空间。

## 1.3.2 策略
有了模型的假设空间，接下来确定选择最优模型的准则，即策略。  
- 损失函数：
    - 0-1损失函数 0-1 loss function:
    - 平方损失函数 quadratic loss function: 
    - 绝对损失函数 absolute loss function: 
    - 对数损失函数 logarithmic loss function 或 对数似然损失函数 log-likelihood loss function
> 损失函数值越小，模型越好。
- 风险函数risk function或期望损失expected loss：模型$f(X)$关于联合分布$P(X,Y)$的平均意义下的损失 $R_{exp}(f)=E_r[L(Y,f(X))]=\int_{x\times y}L(y,f(x))P(x,y)\d x\d y$
> 学习的目标就是选择**期望风险**最小的模型。由于联合分布$P(X,Y)$是未知的，$R_{exp}(f)$不能直接计算，只能通过学习得到。而根据期望风险最小学习模型要用到联合分布，联合分布又未知，这就导致了监督学习的病态ill-formed problem。
